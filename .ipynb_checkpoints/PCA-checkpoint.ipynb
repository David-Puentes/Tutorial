# PCA

El análisis de componentes principales o PCA por sus sigla en inglés es un método multivariante para reducir la dimensionalidad de los datos. Tiene como requisito que las variables deben estar relacionadas entre ellas, de tal manera que se puedan agrupar en componentes.

```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

Importar los datos de "Iris" y eliminar los faltantes

```python
data=pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data",header=None,prefix="X")
data.columns=['sepal_len', 'sepal_wid', 'petal_len', 'petal_wid', 'label']
data.isnull().sum() # Saber si hay casos con NA
data.dropna(how="all", inplace=True) # Eliminar las observaciones con NA
data.columns
```

Separar las variables dependientes de las independientes y normalizarlas

```python
X=data.drop("label",axis = 1)
y=data["label"]
names=data.columns
X_std= StandardScaler().fit_transform(X)
```

```python
plt.figure("Histograma")
data.plot.hist(alpha=0.4)
plt.figure("Gráfico de pie",figsize=(5,5))
data.label.value_counts().plot(kind="pie",autopct='%1.1f%%', fontsize=15)
plt.show()
```

```python
categories = data.label
colors = np.linspace(1, 2, len(categories))
colordict = dict(zip(categories, colors))  
color= data['label'].apply(lambda x: colordict[x])
```

```python
plt.figure(figsize=(15,15))
pd.plotting.scatter_matrix(data,figsize=(15,15),c=color)
plt.show()
```

Construcción de las matrices de varianza y covarianza

```python
# Matriz de covarianza
mean_vec=np.mean(X_std,axis=0)
cov_mat=(X_std-mean_vec).T.dot((X_std-mean_vec))/(X_std.shape[0]-1)
```

```python
#Gráficar en un mapa de calor las covarianzas

corMat=np.asarray(cov_mat)

fig, ax = plt.subplots()
im = ax.imshow(corMat,cmap="RdBu")

# Mostrar las etiquetas de las variables
ax.set_xticks(np.arange(len(names)-1))
ax.set_yticks(np.arange(len(names)-1))
# Asignar los nombres a las etiquetas
ax.set_xticklabels(names[:4])
ax.set_yticklabels(names[:4])

# Rotar los nombres
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

# Loop para asignar los valores de las correlaciones en el mapa de calor
for i in range(len(names)-1):
    for j in range(len(names)-1):
        text = ax.text(j, i, round(corMat[i, j],2),
                       ha="center", va="center", color="w")

plt.title("Matriz de covarianzas", fontdict={"fontsize":15}) # o de correlación si los datos están normalizados
plt.show()
```
