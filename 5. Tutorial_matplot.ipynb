###  <a id='seccion 2'></a>2. Creación de gráficas

```python
# El primer paso para la creación de las gráficas es crear una figura y los ejes, así:
fig = plt.figure()
ax= plt.axes()
```

```python
# Gráfica de una función seno

fig = plt.figure()
ax= plt.axes()

x= np.linspace(0,10,1000)
plt.plot(x,np.sin(x))

# Guardar una imagen
# Para guardar una imagen se usa la función savefig

plt.savefig("figura.png")
```

 Múltiples gráficas

```python
# Alternativa 1

fig,ax=plt.subplots(2)

ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));
```

```python
# Alternativa 2

plt.figure()  # crear una figura

# Crear una gráfica en el primero de dos paneles
plt.subplot(2, 1, 1) # (filas, columnas, número de paneles)
plt.plot(x, np.sin(x))

# Crear una gráfica en el primero de dos paneles
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x));
```

```python
# Dos gráficos en uno solo

plt.axes()
plt.plot(np.linspace(0,10,100),np.exp(np.linspace(0,10,100)))
plt.grid(c="0.5",linestyle="--",fillstyle="top")

# Agregar nuevos ejes al mismo gráfico
plt.axes([0.2,0.5,0.3,0.3],facecolor="0.9") # Se especifíca la posición de los nuevos ejes  
plt.plot(np.linspace(0,10,120),np.sin(np.linspace(0,10,120)),linewidth=1,color="red")
plt.title(" Función Seno");
# plt.box() Quitar "caja"
# plt.delaxes() Elimina el gráfico
# plt.cla() Borrar el contenido del gráfico
```

###  <a id='seccion 3'></a>3. Configuración de la gráfica

Una de las primeras configuraciones que se suele realizar es la del tipo y color de las líneas. En el momento que se crea el gráfico "plot" se especifica el color con la paralabra "color" y el tipo de línea con "linestyle"

```python
?plt.plot()
```

```python
# Color de las líneas

plt.plot(x, np.sin(x - 0), color='blue')        # nombre del color
plt.plot(x, np.sin(x - 1), color='g')           # código corto de color (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # Escala de grises entre 0 y 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # tupla RGB, valores de 0 a 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # colores soportados por HTLM
```

```python
# Tipo de línea

plt.plot(x, x + 4, linestyle='-')  # solida
plt.plot(x, x + 5, linestyle='--') # discontinua
plt.plot(x, x + 6, linestyle='-.') # guion y punto
plt.plot(x, x + 7, linestyle=':'); # punteada

# Si no se especifican los colores matplot lo hace por defecto
```

###  <a id='seccion 4'></a>4. Margenes y etiquetas

 Margenes
 
 Para especificar las margenes de los gráficos se emplean las funciones plt.xlim () y plt.ylim() para los respectivos ejes. De igual manera, se puede emplear la función plt.axis([ ]) en la se hace la configuración directamente.

```python
# Alternativa 1

plt.plot(x,np.sin(x))
plt.xlim(-1,11)
plt.ylim(-1.5,1.5);
```

```python
# Alternativa 2

plt.plot(x,np.sin(x))
plt.axis([-1,11,-1.5,1.5]); # [xmin, xmax, ymin, ymax]
```

Etiquetas

Dentro de las etiquetas se incluye: títulos, etiquetas de ejes y etiquetas de lineas

```python
plt.plot(x,np.sin(x),label="Función 1")
plt.axis([-1,11,-1.5,1.5]) # [xmin, xmax, ymin, ymax]
plt.title("Seno de x")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.legend();
```

```python
plt.figure("Valores por defecto")
plt.plot([1,2,3,4,5],label="Por defecto")
plt.legend(loc=2)

plt.figure("Valores modificados")
plt.rc("lines", linewidth=3) # Modifica el tamaño de las líneas para todos los gráficos de la sesión
plt.rc("font",size=16) # Modifica el tamaño de la fuente para todos los gráficos de la sesión

plt.plot([1,2,3,4,5],label="Modificado")
plt.legend(loc=2);
```

###  <a id='seccion 5'></a>5. Gráficos de dispersión

Otro tipo de gráfico de gran utilidad son los de dispersión. Para la realización de estos se puede acudir a la función scatter o declarar el tipo de línea, que en este caso serían puntos

```python
# Alternativa 1: Función scatter

y=np.sin(np.linspace(0,5,50))
x=np.linspace(0,5,50)

plt.figure("Scatter") # Se genera una figura llamada Scatter
plt.title("Puntos")
plt.scatter(x,y,alpha=0.5);
```

```python
# Alternativa 2: Tipo de línea

y=np.sin(np.linspace(0,5,50))

plt.plot(y,"x",label="Normal")
plt.plot(0.5*y,".",label="Menor")
plt.plot(1.5*y,"2",label="Mayor")
plt.axhline(0,linestyle="--",linewidth=0.5)
plt.legend();
```

```python
dias=[np.array(calendar.mdays)[0:i].sum()+1 for i in np.arange(12)+1]
meses=calendar.month_name[1:13]
a=np.random.rand(365)

plt.axes([0.1,0.2,0.8,0.65])
plt.plot(np.arange(1,366),a,label="Valores")
plt.xlim(-5,370)
plt.ylim(0,1.2)
plt.legend()
plt.title("Gráfica")
plt.suptitle("De nuevo") # Título superior
plt.minorticks_on() # Ver marcadores entre los valores de x y y
plt.axhline(max(a),linestyle="--",linewidth=0.5,c="darkblue")
plt.xticks(dias,meses,size="small",color="black",rotation=45)
plt.xlabel("t (días)")
plt.ylabel("Media diaria");
```

Tipos de gráfico (I)

```python
x=np.arange(100)
y=np.random.rand(100)
plt.plot(x,y,c="black",label="(x,f(x))",linewidth=0.8)
plt.minorticks_on()

#Destacar los valores mayores a 0.9
plt.plot(x[y>0.9],y[y>0.9],"o",label="f(x)>0.9")
plt.axhspan(0.9,1,alpha=0.2)
plt.title("Representación de (x,f(x))")
plt.xlabel("Valores de x")
plt.ylabel("Valores de y")
plt.legend()
```

```python
# Se grafica una seria de tiempo, donde las líneas punteadas son las desviaciones
x=np.arange(25)+1
y=np.random.rand(25)*10
y_norm=(y- np.mean(y))/ np.std(y)

plt.xlim(np.min(x)-1,np.max(x)+1)
plt.ylim(np.min(y_norm)-1,np.max(y_norm)+1)

plt.stem(x[y_norm>0],
        y_norm[y_norm>0],
        linefmt="k-.",
        markerfmt="go",
        basefmt="b-")
plt.stem(x[y_norm<0],
        y_norm[y_norm<0],
        linefmt="k-.",
        markerfmt="ro",
        basefmt="b-")
```

```python
# Crear polígonos cuando los valores de una serie sean mayores a la otra serie
x= np.arange(25)+1
y_1=np.random.rand(25)*10
y_2=np.random.rand(25)*10

plt.plot(x,y_1,color="green",linewidth=1,label="serie 1")
plt.plot(x,y_2,"k-.",linewidth=1,label="serie 2")

plt.fill_between(x,y_1,y_2,where=(y_1>y_2),interpolate=True)
plt.title("Serie de tiempo")
plt.legend();
```

```python
y=np.random.randn(20)
plt.plot(y)
plt.plot(np.argmin(y),np.min(y),"x")
plt.plot(np.argmax(y),np.max(y),"x")
plt.minorticks_on()
plt.text(np.argmin(y),np.min(y)-0.1,u"Mínimo",fontsize=10,
        horizontalalignment="center",verticalalignment="center")
plt.text(np.argmax(y),np.max(y)-0.1,u"Máximo",fontsize=10,
        horizontalalignment="center",verticalalignment="center")
```

```python
#Insertar flechas al gráfico
plt.plot(y)
plt.ylim(-3,3)
plt.annotate(u"Máximo",xy=(np.argmax(y),np.max(y)),xycoords="data",
            xytext=(np.argmax(y)-1,np.max(y)+0.5),
            textcoords="data",
            arrowprops=dict(arrowstyle="->"))
plt.annotate(u"Mínimo",xy=(np.argmin(y),np.min(y)),xycoords="data",
            xytext=(np.argmin(y)-1.7,np.min(y)-0.5),
            textcoords="data",
            arrowprops=dict(arrowstyle="->"));
```

```python
# Insertar tablas a las gráficas

valores=[[np.argmax(y),np.max(y)],[np.argmin(y),np.min(y)]]
label_row=("x","y")
label_col=("Mínimo","Máximo")

plt.plot(y)
plt.ylim(-3,3)
plt.table(cellText=valores,rowLabels=label_row,
          colLabels=label_col,
         colWidths=[0.2]*len(y),loc="upper center");
```

```python
np.random.seed(19680801)

y=np.random.normal(0.5,0.4,1000)
y=y[(y>0)&(y<1)]
y.sort()
x=np.arange(len(y))


# Lineal
plt.subplot(2,2,1)
plt.plot(x,y)
plt.yscale("linear")
plt.title("Lineal")

# logarítmica
plt.subplot(2,2,2)
plt.plot(x,y)
plt.yscale("log")
plt.title("Logaritmo")

#Simétrica

plt.subplot(2,2,3)
plt.plot(x,y-y.mean())
plt.yscale("symlog",linthreshy=0.01)
plt.title("Simétrica")

#Logística
plt.subplot(2,2,4)
plt.plot(x,y)
plt.yscale("logit")
plt.title("Logit")

# Reducir las etiquetas de los ejes
plt.gca().yaxis.set_minor_formatter(NullFormatter())

plt.subplots_adjust(top=0.92,bottom=0.08,left=0.1,right=0.95,
                   hspace=0.35,wspace=0.35)
```
