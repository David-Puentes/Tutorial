```python
x=np.arange(100)
y=np.random.rand(100)
plt.plot(x,y,c="black",label="(x,f(x))",linewidth=0.8)
plt.minorticks_on()

#Destacar los valores mayores a 0.9
plt.plot(x[y>0.9],y[y>0.9],"o",label="f(x)>0.9")
plt.axhspan(0.9,1,alpha=0.2)
plt.title("Representaci√≥n de (x,f(x))")
plt.xlabel("Valores de x")
plt.ylabel("Valores de y")
plt.legend()
```

```python
# Se grafica una seria de tiempo, donde las l√≠neas punteadas son las desviaciones
x=np.arange(25)+1
y=np.random.rand(25)*10
y_norm=(y- np.mean(y))/ np.std(y)

plt.xlim(np.min(x)-1,np.max(x)+1)
plt.ylim(np.min(y_norm)-1,np.max(y_norm)+1)

plt.stem(x[y_norm>0],
        y_norm[y_norm>0],
        linefmt="k-.",
        markerfmt="go",
        basefmt="b-")
plt.stem(x[y_norm<0],
        y_norm[y_norm<0],
        linefmt="k-.",
        markerfmt="ro",
        basefmt="b-")
```

```python
# Crear pol√≠gonos cuando los valores de una serie sean mayores a la otra serie
x= np.arange(25)+1
y_1=np.random.rand(25)*10
y_2=np.random.rand(25)*10

plt.plot(x,y_1,color="green",linewidth=1,label="serie 1")
plt.plot(x,y_2,"k-.",linewidth=1,label="serie 2")

plt.fill_between(x,y_1,y_2,where=(y_1>y_2),interpolate=True)
plt.title("Serie de tiempo")
plt.legend()
```

```python
y=np.random.randn(20)
plt.plot(y)
plt.plot(np.argmin(y),np.min(y),"x")
plt.plot(np.argmax(y),np.max(y),"x")
plt.minorticks_on()
plt.text(np.argmin(y),np.min(y)-0.1,u"M√≠nimo",fontsize=10,
        horizontalalignment="center",verticalalignment="center")
plt.text(np.argmax(y),np.max(y)-0.1,u"M√°ximo",fontsize=10,
        horizontalalignment="center",verticalalignment="center")
```

```python
#Insertar flechas al gr√°fico
plt.plot(y)
plt.ylim(-3,3)
plt.annotate(u"M√°ximo",xy=(np.argmax(y),np.max(y)),xycoords="data",
            xytext=(np.argmax(y)-1,np.max(y)+0.5),
            textcoords="data",
            arrowprops=dict(arrowstyle="->"))
plt.annotate(u"M√≠nimo",xy=(np.argmin(y),np.min(y)),xycoords="data",
            xytext=(np.argmin(y)-1.7,np.min(y)-0.5),
            textcoords="data",
            arrowprops=dict(arrowstyle="->"))
```

```python
# Insertar tablas a las gr√°ficas
valores=[[np.argmax(y),np.max(y)],[np.argmin(y),np.min(y)]]
label_row=("x","y")
label_col=("M√≠nimo","M√°ximo")

plt.plot(y)
plt.ylim(-3,3)
plt.table(cellText=valores,rowLabels=label_row,
          colLabels=label_col,
         colWidths=[0.2]*len(y),loc="upper center")
```

```python
np.random.seed(19680801)

y=np.random.normal(0.5,0.4,1000)
y=y[(y>0)&(y<1)]
y.sort()
x=np.arange(len(y))


# Lineal
plt.subplot(2,2,1)
plt.plot(x,y)
plt.yscale("linear")
plt.title("Lineal")

# logar√≠tmica
plt.subplot(2,2,2)
plt.plot(x,y)
plt.yscale("log")
plt.title("Logaritmo")

#Sim√©trica

plt.subplot(2,2,3)
plt.plot(x,y-y.mean())
plt.yscale("symlog",linthreshy=0.01)
plt.title("Sim√©trica")

#Log√≠stica
plt.subplot(2,2,4)
plt.plot(x,y)
plt.yscale("logit")
plt.title("Logit")

# Reducir las etiquetas de los ejes
plt.gca().yaxis.set_minor_formatter(NullFormatter())

plt.subplots_adjust(top=0.92,bottom=0.08,left=0.1,right=0.95,
                   hspace=0.35,wspace=0.35)
```
