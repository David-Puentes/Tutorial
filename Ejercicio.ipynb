## 1. Medidas de tendencia central

Media

```python
# Aletrnativa 1
media=0
for i in data['sepal_wid']:
    media+=i/len(data)
print(media)
```

```python
# Alternativa 2
media=sum(data['sepal_wid'])/len(data)
print(media)
```

```python
# Alternativa 3
media=data['sepal_wid'].mean()
print(media)
```

Mediana

```python
# Alternativa 1

sepal_ord=sorted(data['sepal_wid'])
n=len(sepal_ord)
central=int(n/2)

if n%2==0:
    mediana=(sepal_ord[central]+sepal_ord[central - 1])/2
else:
    mediana= sepal_ord[central]
print(mediana)
```

```python
# Alternativa 2

mediana=np.median(data['sepal_wid'])
print(mediana)
```

```python
# Alternativa 3
mediana=data['sepal_wid'].median()
print(mediana)
```

Moda

```python
moda=data['sepal_wid'].mode()
print(moda)
```

## 2. Medidas de dispersión

Desviación estándar

```python
Dest=data["sepal_wid"].std()
print(Dest)
```

Rango intercuartil

```python
rango=data["sepal_wid"].max() - data["sepal_wid"].min()
r_inter= data["sepal_wid"].quantile(0.75)-data["sepal_wid"].quantile(0.5)
print(rango,"\n",r_inter)
```

Coeficiente de variación

```python
import scipy.stats as ss
```

```python
cv=data["sepal_wid"].std()/data["sepal_wid"].mean()
cv2=ss.variation(data["sepal_wid"])
print(cv,cv2)
```
